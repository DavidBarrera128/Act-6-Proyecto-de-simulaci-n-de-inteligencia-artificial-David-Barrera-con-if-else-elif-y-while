### Act 6 Proyecto de simulación de inteligencia artificial con if, else, elif y while

```python
# Simulador de Asistente Bancario Virtual - Gestión de Cheques
print("====================================================")
print("🤖 IA BANCO VIRTUAL: SISTEMA DE VERIFICACIÓN DE CHEQUES")
print("====================================================")
print("Hola, soy tu asistente virtual. Te ayudaré a gestionar tus cheques.")

# Configuración inicial simulada
saldo_cuenta_banco = 5000.0
continuar = "si"

# Ciclo principal interactivo
while continuar == "si":
    print("\n--- MENÚ DE OPCIONES ---")
    print("1. Cobrar un cheque")
    print("2. Depositar un cheque")
    print("3. Consultar fondos de la cuenta")
    print("4. Salir")

    opcion = input("\nPor favor, elige una opción (1-4): ")

    # Opción 1: Cobrar un Cheque
    if opcion == "1":
        print("\n[Procesando Solicitud de Cobro]")
        monto_cheque_str = input("Introduce el monto del cheque a cobrar ($): ")
        monto_cheque = float(monto_cheque_str)

        if monto_cheque <= 0:
            print("❌ Error de IA: El monto del cheque debe ser mayor a 0.")
        elif monto_cheque > saldo_cuenta_banco:
            print("❌ Fondos Insuficientes: El cheque excede el saldo disponible.")
            print("🤖 Sugerencia de IA: Intenta depositar fondos primero o solicita un cheque menor.")
        else:
            saldo_cuenta_banco = saldo_cuenta_banco - monto_cheque
            print("✅ Cheque procesado con éxito.")
            print("Retiraste: $", monto_cheque)
            print("Saldo restante en cuenta: $", saldo_cuenta_banco)

    # Opción 2: Depositar un Cheque
    elif opcion == "2":
        print("\n[Procesando Depósito de Cheque]")
        monto_deposito_str = input("Introduce el monto del cheque a depositar ($): ")
        monto_deposito = float(monto_deposito_str)

        if monto_deposito <= 0:
            print("❌ Error de IA: No puedes depositar un cheque sin valor o negativo.")
        elif monto_deposito > 10000:
            print("⚠️ Alerta de Seguridad de IA: Los depósitos mayores a $10,000 requieren aprobación.")
            print("El cheque entrará en periodo de retención de 24 horas.")
        else:
            saldo_cuenta_banco = saldo_cuenta_banco + monto_deposito
            print("✅ Depósito exitoso.")
            print("Nuevo saldo disponible: $", saldo_cuenta_banco)

    # Opción 3: Consultar Fondos
    elif opcion == "3":
        print("\n[Consultando Estado de Cuenta]")
        print("🤖 Análisis de IA: Tu salud financiera actual es estable.")
        print("Saldo disponible para respaldar cheques: $", saldo_cuenta_banco)

        if saldo_cuenta_banco < 1000:
            print("⚠️ Advertencia: Tu saldo es bajo. Evita emitir cheques grandes.")

    # Opción 4: Salir del programa
    elif opcion == "4":
        print("\n🤖 IA: Gracias por utilizar el sistema de cheques virtual. ¡Hasta pronto!")
        continuar = "no"

    # Opción No válida
    else:
        print("❌ Opción no válida. Por favor, selecciona un número del 1 al 4.")

    # Preguntar si desea continuar si no eligió salir formalmente
    if opcion != "4":
        continuar = input("\n¿Deseas realizar otra operación? (si/no): ")
        continuar = continuar.lower() # Convierte a minúsculas por si escriben "SI"

        if continuar != "si":
            print("\n🤖 IA: Sesión finalizada. Que tengas un excelente día.")

print("\n====================================================")
print("Programa terminado.")
print("====================================================")
print("Programa realizado por David Barrera 1388")
```
